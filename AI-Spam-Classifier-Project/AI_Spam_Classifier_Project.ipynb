```python
# 1. Split the data
# X = messages
# y = labels (spam or ham)

X_train, X_test, y_train, y_test = train_test_split(
    df['text'],
    df['label'],
    test_size=0.2,
    random_state=60

)

# 2. Create the CountVectorizer
vectorizer = CountVectorizer()

# 3. Convert the training messages into numbers
X_train_math = vectorizer.fit_transform(X_train)

# 4. Convert the testing messages into numbers
X_test_math = vectorizer.transform(X_test)

print("Words successfully converted to numbers!")
```

............................

```python
# Initialize the Naive Bayes AI model
ai_classifier = MultinomialNB()

# Train the AI model
ai_classifier.fit(X_train_math, y_train)

print("The AI has successfully trained itself!")
```

.....................

```python
# Ask the AI to predict the answers for our hidden test set
predictions = ai_classifier.predict(X_test_math)

# Calculate the accuracy
accuracy = accuracy_score(y_test, predictions)

# Display the accuracy
print(f"AI Accuracy Score: {accuracy * 100}%\n")

# Display a detailed report
print("Detailed Report Card:")
print(classification_report(y_test, predictions))
```

...............................

```python
def check_spam(message):
    # Convert the message into numbers
    math_message = vectorizer.transform([message])

    # Predict whether it is spam or ham
    prediction = ai_classifier.predict(math_message)[0]

    # Display the result
    if prediction == 'spam':
        print(f"SPAM DETECTED: '{message}'")
    else:
        print(f"SAFE (Ham): '{message}'")
```
